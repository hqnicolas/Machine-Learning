```python
# Deletando as Colunas
dataset.drop(['Embarked'], axis=1, inplace = True)

# Verificando os dados faltantes.
dataset.isna().sum()

# Substituindo os dados faltantes pela mdia.
dataset.fillna(dataset.mean(), inplace=True)

# Droppando as colunas que no foram ajustadas.
dataset.dropna(axis=1, inplace=True)
```

```python
# Deletando as informaes que devem estar no modelo.
dataset.drop(['PassengerId'], axis=1, inplace=True)
dataset
```

```python
# Pegando todas as variveis object e bool
# para aplicar a transformao ordinal encoder.
columns = dataset.select_dtypes(include=['object', 'bool']).columns
columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal = OrdinalEncoder()
```

```python
values_cat = ordinal.fit_transform(dataset[columns])
```

```python
dataset[columns] = values_cat
dataset
```

```python
x = dataset.drop(['Survived'], axis=1)
y = dataset['Survived']
```

```python
# Realizando a separao de x e y em dados de treino (70%) e teste (30%).
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
# Importando o modelo para regresso logistic.
from sklearn.linear_model import LogisticRegression

# Instanciando o modelo de regresso.
clf_Logistic = LogisticRegression(random_state=0)

# Treinando o modelo de regresso.
clf_Logistic.fit(X_train, y_train)
```

```python
# Realizando a predio nos dados de teste.
# Pega as informaes das variveis independentes (somente teste).
y_predict = clf_Logistic.predict(X_test)

# Importando a lib para avaliar o modelo (acurcia)
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict)
```
